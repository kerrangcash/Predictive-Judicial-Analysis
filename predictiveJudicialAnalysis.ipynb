```python
#fields present in JSON dataset of cases
ITEMID = data.Field()
#LANGUAGEISOCODE = data.Field()
RESPONDENT = data.Field()
BRANCH = data.Field()
#DATE = data.Field()
#DOCNAME = data.Field()
IMPORTANCE = data.Field()
CONCLUSION = data.Field()
JUDGES = data.Field()
TEXT = data.Field()
#VIOLATED_ARTICLES = data.Field()
#VIOLATED_PARAGRAPHS = data.Field()
#VIOLATED_BULLETPOINTS = data.Field()
#NON_VIOLATED_ARTICLES = data.Field()
#NON_VIOLATED_PARAGRAPHS = data.Field()
#NON_VIOLATED_BULLETPOINTS = data.Field()

#dictionary of fields for use in batching
fields = {'ITEMID': ('id', ITEMID), 
          #'LANGUAGEISOCODE': ('lan', LANGUAGEISOCODE), 
          'RESPONDENT': ('res', RESPONDENT), 
          'BRANCH': ('bran', BRANCH), 
          #'DATE': ('da', DATE), 
          #'DOCNAME': ('doc', DOCNAME),
          'IMPORTANCE': ('imp', IMPORTANCE), 
          'CONCLUSION': ('con', CONCLUSION), 
          'JUDGES': ('jdg', JUDGES),
          'TEXT': ('txt', TEXT), 
          #'VIOLATED_ARTICLES': ('vla', VIOLATED_ARTICLES), 
          #'VIOLATED_PARAGRAPHS': ('vlp', VIOLATED_PARAGRAPHS),
          #'VIOLATED_BULLETPOINTS': ('vlb', VIOLATED_BULLETPOINTS), 
          #'NON_VIOLATED_ARTICLES': ('nvla', NON_VIOLATED_ARTICLES), 
          #'NON_VIOLATED_PARAGRAPHS': ('nvlp', NON_VIOLATED_PARAGRAPHS),
          #'NON_VIOLATED_BULLETPOINTS': ('nvlb', NON_VIOLATED_BULLETPOINTS)
          }
```

```python
#creating datasets
train_data, test_data = data.TabularDataset.splits(
                            #path = 'dataset\ECHR_dataset',
                            #train = r"EN_train\001-4496.json",
                            path = '',
                            train = 'exampleTrain.json',
                            test = 'example.json',
                            format = 'json',
                            fields = fields
)
```

```python
print(vars(train_data[1]))
print(vars(test_data[2]))
```

```python
import torch
```

```python
ITEMID.build_vocab(train_data)
RESPONDENT.build_vocab(train_data)
BRANCH.build_vocab(train_data)
#DATE.build_vocab(train_data)
IMPORTANCE.build_vocab(train_data)
CONCLUSION.build_vocab(train_data)
JUDGES.build_vocab(train_data)
TEXT.build_vocab(train_data)
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

BATCH_SIZE = 2

train_iterator, test_iterator = data.BucketIterator.splits(
    (train_data, test_data),
    sort = False, #don't sort test/validation data
    batch_size=BATCH_SIZE,
    device=device)
```

```python
print('Train:')
for batch in train_iterator:
    print(batch)
```
